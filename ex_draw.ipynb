```python
#--------------------------------------------数据拟合------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# 设置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# 假设DataFrame df已经定义好并且包含'Flow', 'Density', 'Speed', '秒'等列
# 读取Excel文件
file_path = '.\output\RepLFA_2024_09_27_09_30_59.csv'  # 替换为你的文件路径
df = pd.read_csv(file_path).iloc[:]
# 创建线性回归模型实例
model = LinearRegression()

# Flow与时间的关系
plt.figure(figsize=(14, 10))

plt.subplot(1, 1, 1)
plt.scatter(df.index, df['untrust_ip_dst_entropy'], label='熵值', alpha=0.6,s=10)
# 拟合线性回归模型
model.fit(df[['time']].values, df['untrust_ip_dst_entropy'].values)
# 计算预测值
frame_fit = np.linspace(df.index.min(), df.index.max(), 100).reshape(-1, 1)
traffic_fit = model.predict(frame_fit)
plt.plot(frame_fit, traffic_fit, label='拟合曲线', color='red')
plt.title('目的地址熵时序图')
plt.xlabel('时间 (秒)')
plt.ylabel('熵')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
file_path = '.\\output\\RepLFA_2024_09_27_16_00_39.csv'  # 注意这里的路径需要根据实际情况调整
df = pd.read_csv(file_path)

# 创建一个新的figure
plt.figure(figsize=(11, 3))

# 选择untrust_ip_dst_entropy大于等于6的数据点并绘制
df_high_entropy = df[df['untrust_ip_dst_entropy'] >= 6]
plt.scatter(df_high_entropy.index, df_high_entropy['untrust_ip_dst_entropy'],  label='normal entropy', alpha=0.6, s=10)

# 选择untrust_ip_dst_entropy小于6的数据点并绘制
df_low_entropy = df[df['untrust_ip_dst_entropy'] < 6]
plt.scatter(df_low_entropy.index, df_low_entropy['untrust_ip_dst_entropy'], color='red', label='abnormal entropy', alpha=0.6, s=10)


plt.title('Timeline of the destination address entropy')
plt.xlabel('Time (s)')
plt.ylabel('Entropy')
plt.legend()

plt.tight_layout()
plt.show()
```
